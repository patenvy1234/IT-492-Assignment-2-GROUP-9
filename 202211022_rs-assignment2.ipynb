```python
import pandas as pd
from scipy.sparse import csr_matrix


# Keep only the required columns
df = df1[['rating', 'gPlusUserId', 'gPlusPlaceId']]

# Drop any rows with missing values
df.dropna(inplace=True)

# Convert the IDs to integer codes
df['user_id'] = pd.factorize(df['gPlusUserId'])[0]
df['place_id'] = pd.factorize(df['gPlusPlaceId'])[0]

# Create a sparse matrix
matrix = csr_matrix((df['rating'].astype(float), (df['user_id'], df['place_id'])))

# Print the matrix
print(matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity between all pairs of columns in the matrix
cosine_sim = cosine_similarity(matrix.T)

# Print the cosine similarity matrix
print(cosine_sim)
```

```python
type(cosine_sim)
```

```python
# Get a list of all unique place IDs in the dataset
unique_place_ids = df['gPlusPlaceId'].unique()

# Create a dictionary mapping each place ID to its index in the matrix
place_id_to_index = {place_id: i for i, place_id in enumerate(unique_place_ids)}
```

```python
index_to_place_id = {v: k for k, v in place_id_to_index.items()}
```

```python
def get_top_similar_places(place_id, k=10):
    # Get the index of the input place ID in the matrix
    place_index = place_id_to_index[place_id]
    
    # Get the cosine similarity scores between the input place and all other places
    similarity_scores = cosine_sim[place_index]
    
    # Get the indices of the top-k similar places
    top_k_indices = similarity_scores.argsort()[::-1][1:k+1]
    
    # Convert the indices back to place IDs
    top_k_places = [index_to_place_id[i] for i in top_k_indices]
    
    return top_k_places
```

```python
place_id = '100555265759855404965'  # Example place ID
top_k_places = get_top_similar_places(place_id, k=5)  # Get top 5 similar places
print(top_k_places)  # Print the top similar places
```

```python
from textblob import TextBlob

# Define a function to perform sentiment analysis on a given text
def get_sentiment(text):
    # Create a TextBlob object for the input text
    blob = TextBlob(text)
    # Get the sentiment polarity of the text (ranging from -1 to 1)
    sentiment_polarity = blob.sentiment.polarity
    # Return the sentiment polarity
    return sentiment_polarity

# Apply the get_sentiment function to the reviewText column
df2 = df1.dropna(subset=['reviewText'])
df2['sentiment'] = df2['reviewText'].apply(get_sentiment)

# Print the first 10 rows of the updated dataframe
print(df2.head(10))
```

```python
import pandas as pd
from scipy.sparse import csr_matrix
from sklearn.feature_extraction.text import TfidfVectorizer
from textblob import TextBlob

# Load the data
# df = pd.read_json('/content/drive/MyDrive/RS/Assignment2-data/reviews.clean.json.gz', lines=True)

# Create a TfidfVectorizer object to extract features from the review text
tfidf = TfidfVectorizer(stop_words='english')

# Compute the tf-idf matrix
tfidf_matrix = tfidf.fit_transform(df1['reviewText'])

# Compute the sentiment score for each review
sentiment_scores = [TextBlob(text).sentiment.polarity for text in df1['reviewText']]

# Assign a weight to each review based on its sentiment score
weights = [score + 1 for score in sentiment_scores]

# Create a sparse matrix with the weighted ratings
matrix = csr_matrix((df1['rating'].astype(float) * weights, (df1['gPlusUserId'], df1['gPlusPlaceId'])))

# Compute the cosine similarity between the places
similarities = cosine_similarity(matrix)

# Define a function to get the top k similar places
def get_top_similar_places_sa(place_id, k=10):
    # Get the index of the input place ID in the matrix
    place_index = place_id_to_index[place_id]

    # Get the cosine similarity scores between the input place and all other places
    place_similarities = similarities[place_index]

    # Get the indices of the top k similar places
    top_k_indices = place_similarities.argsort()[::-1][:k]

    # Convert the indices back to place IDs
    top_k_places = [index_to_place_id[i] for i in top_k_indices]

    return top_k_places

# Define a function to get the name of a place given its ID
# def get_place_name(place_id):
#     return df.loc[df1['gPlusPlaceId'] == place_id, 'name'].iloc[0]
```
