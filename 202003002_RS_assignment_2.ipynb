#Data loading

```python
N = 2000000 # getting the first 20 lakh rows 
data_reviews = []
with gzip.open('/content/drive/My Drive/reviews.clean.json.gz') as f:
    head = [next(f) for x in range(N)]
    for l in head:
        data_reviews.append(eval(l))
```

```python
df_data_reviews = pd.DataFrame.from_dict(data_reviews)
```

```python
df_data_reviews = df_data_reviews.explode('categories')
df_data_reviews = df_data_reviews.dropna(subset = ["categories"])
```

```python
df_data_reviews = df_data_reviews.loc[df_data_reviews["categories"].str.endswith('Restaurant')]  # considering only those reviews which are rated to a restruant
```

```python
filter_df = df_data_reviews[df_data_reviews.groupby("gPlusUserId")["categories"].transform("size")>10]  # considering only those users who have rated greater then 10 times
final_df = filter_df[filter_df.groupby("categories")["categories"].transform("size")>3]  ## considering only those restuarant types which to which atleast 3 users have rated
```

```python
def count_words(s):
    return len(s.split())
```

```python
final_df["temp"] = final_df.apply(lambda x: count_words(x['categories']),axis=1)
final_df = final_df.loc[final_df['temp'] > 1]
final_df = final_df.drop(['temp'],axis=1)
```

```python
final_df
```

#USER-USER

# creating user-item matrix

```python
user_place_mat = final_df.pivot_table(index="gPlusUserId",columns="categories",values="rating")
user_place_mat
```

```python
meanmapper = user_place_mat
meanmapper["meaner"] = meanmapper.mean(axis=1)  ## this data frame act as dictionary by storing mean rating of each user and its user id
meanmapper = meanmapper[["meaner"]]
```

# mean centring 

```python
user_place_mat = user_place_mat.subtract(user_place_mat.mean(axis=1),axis=0)  ## as each user rates differently the mean centring is done
```

```python
user_place_mat
```

```python
normed = user_place_mat.subtract(user_place_mat.mean(axis=1),axis=0)   ## this is the same mean centred matrix but this will help us find pearson correlation
normed = normed.fillna(0) 
```

```python
meanmapper
```

```python
normed
```

# adjusted cosine 

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sim_mat_user = cosine_similarity(normed)  ##  as normed matrix is already mean centred the cosine simillarity will adjusted cosine
sim_mat_user = pd.DataFrame(sim_mat_user,columns=normed.index.values,index = normed.index.values) 
```

# predicting rating for a user 

```python
user = str(100000524810171549476)
rem_sim_mat = sim_mat_user.drop(index=user) #dropping the particular picked user from the similarity matrix 
n=5 # number of neighbours
simusers = rem_sim_mat[user].sort_values(ascending=False)[:n] # getting the similar user for given user using the filtered sim matrix
print("similar users for ",user," are")
print(simusers)
```

```python
rem_sim_mat
```

```python
already_watch = user_place_mat[user_place_mat.index==user].dropna(axis=1,how='all')  ## geting out restraunts that user has already rated
```

```python
already_watch
```

```python
simi_user_seen = user_place_mat[user_place_mat.index.isin(simusers.index)].dropna(axis=1,how='all') # removing all such restruants that none of the neighbours has rated
```

```python
simi_user_seen
```

```python
simi_user_seen.drop(already_watch.columns,axis=1,inplace=True,errors="ignore") # removing already rated restruants
```

```python
simi_user_seen
```

# Getting recommendations

```python
item_score = {}
for i in simi_user_seen.columns:  # for each restruant in the final filtered mat
  score = meanmapper.loc[user]["meaner"] # adding the mean back as we have subtracted it earlier
  movie_rating = simi_user_seen[i]
  for u in simusers.index:
    if pd.isna(movie_rating[u])==False:
      number_of_common_rated = (user_place_mat.loc[user].notnull() & user_place_mat.loc[u].notnull()).sum()
      if(number_of_common_rated > 5):      ## sim_score
          number_of_common_rated = 5
      score = score + sim_mat_user.loc[user][u]*movie_rating[u]*number_of_common_rated/abs(sim_mat_user.loc[user][u]*number_of_common_rated) # weighted mean centered calculation of rating
  item_score[i] = score
```

```python
final_recommendations = sorted(item_score.items(), key=lambda x:x[1],reverse=True)
```

```python
final_recommendations[:10] ## get the top 10 recommendations
```

#ITEM-ITEM

```python
item_user = final_df.pivot_table(index="categories",columns="gPlusUserId",values="rating")
item_user
```

```python
normed_nan = item_user.subtract(item_user.mean(axis=1),axis=0)
```

```python
normed_nan
```

```python
normed_fill = normed_nan.fillna(0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
item_sim = cosine_similarity(normed_fill)  ##  as normed matrix is already mean centred the cosine simillarity will produce adjusted cosine
item_sim = pd.DataFrame(item_sim,columns=normed_fill.index.values,index = normed_fill.index.values)  # item-item similarity matrix
```

```python
item_sim
```

```python
user_id = str(100000524810171549476) # random user
user_watched = pd.DataFrame(item_user[user_id].dropna(axis=0,how="all").sort_values(ascending=False).reset_index().rename(columns={user_id:"rating"})) ## getting all restruant already rated by user
```

```python
user_watched
```

```python
unkn_rest = "African Restaurant" # setting a restuarant type not visited by above define user
```

```python
unkn_rest_sim = item_sim[[unkn_rest]].reset_index().rename(columns={"index":"categories","African Restaurant": "sim_score"})
```

```python
unkn_rest_sim #Unknown restuarant type similarity with every other restuarant type
```

```python
user_previous_rated_sim = pd.merge(left = user_watched,right=unkn_rest_sim,on="categories",how="inner").sort_values("sim_score",ascending=False)[:5]
```

```python
user_previous_rated_sim # merging both user rated and similarity tables and getting top 5 similar restuarant that user has already rated
```

```python
predicted_rat = round(np.average(user_previous_rated_sim["rating"],weights=user_previous_rated_sim["sim_score"]),5) # average rating for this top 5 restuarant with the unkown restuarant
```

```python
unrated_res = pd.DataFrame(item_user[user_id].isna()).reset_index()
unrated_res = unrated_res[unrated_res[user_id]==True]["categories"].values.tolist()
```

```python
rating_preder = {}
```

```python
for unrate in unrated_res:
  unkn_rest_sim = item_sim[[unrate]].reset_index().rename(columns={"index":"categories",unrate: "sim_score"})
  user_previous_rated_sim = pd.merge(left = user_watched,right=unkn_rest_sim,on="categories",how="inner").sort_values("sim_score",ascending=False)[:5]
  predicted_rat = round(np.average(user_previous_rated_sim["rating"],weights=user_previous_rated_sim["sim_score"]),5) # average rating for this top 5 restuarant with the unkown restuarant
  rating_preder[unrate] = predicted_rat
```

```python
rating_preder = sorted(rating_preder.items(), key=lambda x:x[1],reverse=True)
```
